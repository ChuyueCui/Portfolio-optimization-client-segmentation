## 4.Clustering Assets <a id="4-Clustering"></a>

### 4.1 Assets Clustering <a id="4.1-Assets-Clustering"></a>

- To better understand the different types of assets in our dataset, we applied clustering based on two key investment features: average annual return and volatility.
- We first used the Elbow Method to figure out how many clusters made sense. As you can see in the graph, the inertia drops sharply when going from 1 to 2 clusters, and then starts to level off. This tells us that 2 clusters is a good balance — we’re capturing meaningful differences without overcomplicating things.
- This clustering helps us think about asset selection in a more structured way: not just by individual performance, but by understanding which type of investment they belong to.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Step 1: Calculate features for clustering
cluster_features = pd.DataFrame({
    'Mean Return': returns_ts.mean(),
    'Volatility': returns_ts.std(),
    'Skewness': returns_ts.skew(),
    'Kurtosis': returns_ts.kurt()
})
display(cluster_features.round(3))
```

```python
# Step 2: Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(cluster_features)

# Step 3: Run KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
cluster_labels = kmeans.fit_predict(X_scaled)

# Add labels back to the DataFrame
cluster_features['Cluster'] = cluster_labels
display(cluster_features.sort_values('Cluster'))


import matplotlib.pyplot as plt
import seaborn as sns

# Plot clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=cluster_features,
    x='Volatility', y='Mean Return',
    hue='Cluster', palette='Set2', s=100
)

for i, label in enumerate(cluster_features.index):
    plt.text(cluster_features.loc[label, 'Volatility'] + 0.001,
             cluster_features.loc[label, 'Mean Return'],
             label, fontsize=9)

plt.title("Asset Clusters Based on Risk & Return")
plt.xlabel("Volatility (Std Dev)")
plt.ylabel("Mean Return")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 4.2 Customer Clustering <a id="4.2-Customer-Clustering"></a>

- The Elbow Method graph visualizes the optimal number of clusters for segmenting clients based on their risk score, investment time horizon, liquidity needs, and net worth.
- Inertia drops sharply until `k=3`, after which the improvement plateaus, indicating 3 clusters is a suitable choice.
- The code standardizes client features and applies KMeans clustering with `k=3` to segment clients.
- Cluster labels are appended to the client profile DataFrame under the column `Client_Segment`.
- A segment summary table shows the average attributes per cluster:
  - Segment 0: Moderate risk, medium time horizon, low liquidity need, highest net worth.
  - Segment 1: High risk, long time horizon, no liquidity need, lowest net worth.
  - Segment 2: Low risk, short time horizon, high liquidity need, moderate net worth.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Encode 'Liquidity_Need'
client_profiles['Liquidity_Need_Code'] = client_profiles['Liquidity_Need'].str.strip().str.lower().map({
    'low': 0,
    'medium': 1,
    'high': 2
})

# Select clustering features
features = ['Risk_Score', 'Time_Horizon', 'Liquidity_Need_Code', 'Net_Worth']
client_data = client_profiles[features].dropna()

# Standardize
scaler = StandardScaler()
scaled_client_data = scaler.fit_transform(client_data)

# Elbow method
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, n_init='auto', random_state=42)
    kmeans.fit(scaled_client_data)
    inertia.append(kmeans.inertia_)

# Plot elbow
import matplotlib.pyplot as plt
plt.plot(range(1, 10), inertia, marker='o')
plt.title('Elbow Method for Client Clustering')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
# Apply KMeans with k=3
kmeans_clients = KMeans(n_clusters=3, n_init='auto',random_state=42)
client_clusters = kmeans_clients.fit_predict(scaled_client_data)

# Add cluster label to original DataFrame
client_profiles['Client_Segment'] = client_clusters

# Group summary for each segment
segment_summary = client_profiles.groupby('Client_Segment')[['Risk_Score', 'Time_Horizon', 'Liquidity_Need_Code', 'Net_Worth']].mean()
print(segment_summary)
```

## 5.Model Implementation <a id="5-Model-Implementation"></a>

### 5.1 Base Model (Markowitz Model) - Min. Variance <a id="5.1-Base-Model-(Markowitz-Model)"></a>

```python
import pandas as pd
import numpy as np
from gurobipy import Model, GRB, QuadExpr

# Load asset return data
asset_returns_df = pd.read_csv("asset_returns.csv")
asset_returns_df['Date'] = pd.to_datetime(asset_returns_df[['Year', 'Month']].assign(DAY=1))
returns_pivot = asset_returns_df.pivot(index='Date', columns='Asset_Name', values='Return')

# Drop columns with missing values
returns_pivot = returns_pivot.dropna(axis=1)
cov_matrix = returns_pivot.cov()
mean_returns = returns_pivot.mean()

assets = list(cov_matrix.columns)
n = len(assets)

# Set target return (mean of all assets)
target_return = mean_returns.mean()

# Gurobi model
m = Model("Markowitz_MinVariance")
m.setParam('OutputFlag', 0)

# Decision variables
w = {i: m.addVar(lb=0, ub=1, name=f"w_{i}") for i in assets}

# Objective: minimize portfolio variance
variance_expr = QuadExpr()
for i in range(n):
    for j in range(n):
        variance_expr.add(w[assets[i]] * w[assets[j]] * cov_matrix.iloc[i, j])
m.setObjective(variance_expr, GRB.MINIMIZE)

# Constraints
m.addConstr(sum(w[i] for i in assets) == 1, "TotalWeight")
m.addConstr(sum(w[i] * mean_returns[i] for i in assets) >= target_return, "MinExpectedReturn")

# Solve
m.optimize()

# Output
print(f"Expected Portfolio Return: {sum(w[i].X * mean_returns[i] for i in assets):.4f}")
print(f"Portfolio Variance: {variance_expr.getValue():.6f}")
print("Selected Weights:")
for i in assets:
    if w[i].X > 0.001:
        print(f"{i}: {w[i].X:.4f}")
```

##### The Markowitz Minimum Variance Portfolio Optimization Model
* Calculates the covariance matrix and mean returns of the assets.
* A target return is defined as the average of all asset means.
* Decision variables representing the allocation weight of each asset, constrained to lie between 0 and 1 and sum to 1.
* Objective function minimizes the portfolio’s overall variance, with the a constraint of minimum expected return.
* The model achieves an expected return of approximately 1.32% with a notably low variance of 0.053%. The optimal weight distribution emphasizes stability, with nearly 49% allocated to US Treasury 10-Year bonds, offering strong downside protection. Complementary positions in Microsoft Corp. - **Model Setup**:  
  - Each asset is represented by a **continuous weight variable** (allocation) and a **binary selection variable**  
  - Linking constraints ensure that **only selected assets can receive nonzero allocation**  
  - The total portfolio allocation must **fully utilize the budget** (weights sum to 1)  
  - A diversification rule requires the portfolio to include **at least 3 different assets**

#### 5.3.4 Foundation Client Portfolio Optimization <a id="5.3.4-Foundation-Client-Portfolio-Optimization"></a>


- **Model Objective**: The optimization model maximizes expected annual return for a foundation client with a 430,000 budget, while adhering to constraints that reflect governance, diversification, and investment structure policies.

- **Key Investment Constraints**:
  - A maximum of 10 total asset positions is allowed.
  - Each selected asset must be allocated at least 1% of the portfolio.
  - Alternative investments (like, Gold, REITs) must be made in 20,000 increments.
  - If a technology asset is selected, its allocation must be at least 8% of the portfolio.
  - The portfolio must include at least 3 different asset classes to ensure diversification.

- **Optimization Behavior**:
  - The model prefers Apple Inc. (S001), allocating over 98% of the budget to it due to its strong expected return.
  - Johnson & Johnson (S003) and JPMorgan Chase (S004) are included with minimal allocations just above the 1% threshold to satisfy the diversification constraint.

- **Constraint Satisfaction**:
  - The solution respects the 3-asset-class minimum through small allocations to healthcare and financials alongside technology.
  - No ghost selections are present—each chosen asset has a non-trivial allocation.

- **Strategic Insight**:
  - While the optimizer clearly prioritizes performance-driven assets, it cleverly meets diversification rules with minimal capital, demonstrating efficiency in constraint navigation without diluting return potential.

```python
from gurobipy import Model, GRB, quicksum

# === Inputs ===
assets = list(asset_df['Asset_ID'])
returns = dict(zip(asset_df['Asset_ID'], asset_df['Avg_Annual_Return']))
classes = dict(zip(asset_df['Asset_ID'], asset_df['Asset_Class']))
is_alt = dict(zip(asset_df['Asset_ID'], asset_df['Alternative']))
is_tech = dict(zip(asset_df['Asset_ID'], asset_df['Technology'])) 

total_budget = 430000

# === Model ===
m = Model("FoundationClientPortfolio")
m.setParam("OutputFlag", 0)

# === Variables ===
x = {i: m.addVar(lb=0, ub=1, vtype=GRB.CONTINUOUS, name=f"x_{i}") for i in assets}  # fraction of total budget
y = {i: m.addVar(vtype=GRB.BINARY, name=f"y_{i}") for i in assets}  # selection indicator
z_class = {}  # indicator for each asset class used

# Unique asset classes
asset_classes = list(set(classes.values()))
for c in asset_classes:
    z_class[c] = m.addVar(vtype=GRB.BINARY, name=f"z_{c}")

# === Objective: Maximize return ===
m.setObjective(quicksum(returns[i] * x[i] * total_budget for i in assets), GRB.MAXIMIZE)

# === Constraints ===

# 1. Budget
m.addConstr(quicksum(x[i] * total_budget for i in assets) <= total_budget, "Budget")

# 2. Max 10 total positions
m.addConstr(quicksum(y[i] for i in assets) <= 10, "MaxPositions")

# 3. Link selection to allocation
for i in assets:
    m.addConstr(x[i] * total_budget >= 0.01 * total_budget * y[i], f"MinAllocIfSelected_{i}")
    m.addConstr(x[i] <= y[i], f"Link_{i}")


# 4. Alternative investments in $20,000 increments
z_alt = {i: m.addVar(vtype=GRB.INTEGER, name=f"z_alt_{i}") for i in assets if is_alt[i]}
for i in z_alt:
    m.addConstr(x[i] * total_budget == 20000 * z_alt[i], f"AltIncrement_{i}")

# 5. Tech stock ≥ 8% if selected
for i in assets:
    if is_tech[i]:
        m.addConstr(x[i] * total_budget >= 0.08 * total_budget * y[i], f"TechMin_{i}")

# 6. Asset class tracking (ensure exact use of classes)
for c in asset_classes:
    assets_in_c = [i for i in assets if classes[i] == c]

    # If any asset in class is selected → z_class[c] = 1
    for i in assets_in_c:
        m.addConstr(y[i] <= z_class[c], f"ClassUse_{i}_{c}")

    # If class used → at least one asset from class selected
    m.addConstr(quicksum(y[i] for i in assets_in_c) >= z_class[c], f"ClassUsedIfSelected_{c}")

# 7. At least 3 different asset classes used
m.addConstr(quicksum(z_class[c] for c in asset_classes) >= 3, "MinAssetClasses")

# === Solve ===
m.optimize()

# === Results ===
if m.status == GRB.OPTIMAL:
    print(f"\nExpected Return: ${m.objVal:.2f}")
    print("Selected Asset Allocations:")
    for i in assets:
        if y[i].X > 0.5:
            asset_name = asset_df.loc[asset_df['Asset_ID'] == i, 'Asset_Name'].values[0]
            alloc = x[i].X * total_budget
            print(f"{asset_name} ({i}): ${alloc:.2f}")
else:
    print("No feasible solution found.")
```

```python
asset_char_df['Is_Stock'] = asset_char_df['Asset_Class'].str.contains('Tech|Healthcare|Financial|Consumer|Equity', case=False)
asset_char_df['Is_Bond'] = asset_char_df['Asset_Class'].str.contains('Bond|Fixed Income', case=False)
asset_char_df['Is_International'] = asset_char_df['Asset_Class'].str.contains('International|Emerging', case=False)
asset_char_df['Is_Alternative'] = asset_char_df['Asset_Class'].str.contains('REIT|Alternative|Commodities', case=False)
asset_char_df['Is_Illiquid'] = asset_char_df['Description'].str.contains('Private|Illiquid', case=False)
asset_char_df['Is_Investment_Grade'] = asset_char_df['Beta'] < 1.0

from pulp import LpVariable, LpProblem, lpSum, LpMaximize, LpBinary, value
investment_constraints = pd.read_csv("investment_constraints.csv")

# Define total investment amount
TOTAL_INVESTMENT = 1_000_000

# Load mean returns (assumed annual return as decimal)
mean_returns = dict(zip(asset_char_df['Asset_Name'], asset_char_df['Avg_Annual_Return']))
print(" Mean returns used:")
for a, r in mean_returns.items():
    print(f"{a}: {r:.4f}")

# Create LP model
model = LpProblem("Maximize_Expected_Dollar_Return", LpMaximize)

# Decision variables
alloc = {a: LpVariable(f"alloc_{a.replace(' ', '_')}", lowBound=0, upBound=TOTAL_INVESTMENT) for a in mean_returns}
select = {a: LpVariable(f"select_{a.replace(' ', '_')}", cat=LpBinary) for a in mean_returns}

# Lot-based integer variables (for enforcing $25K lots)
z = {a: LpVariable(f"lots_{a.replace(' ', '_')}", lowBound=0, cat="Integer") for a in mean_returns}

# Enforce allocation = $25,000 * z for all assets
for a in mean_returns:
    model += alloc[a] == 25_000 * z[a], f"LotSize_{a}"

# Objective: maximize expected dollar return
model += lpSum([mean_returns[a] * alloc[a] for a in mean_returns]), "Total_Dollar_Return"

# Budget: total investment = $1M
model += lpSum(alloc[a] for a in mean_returns) == TOTAL_INVESTMENT, "TotalBudget"

# Link selection to allocation
for a in mean_returns:
    model += alloc[a] <= TOTAL_INVESTMENT * select[a], f"LinkSelect_{a}"

# Max 5 stock positions
stock_assets = ['Apple Inc.', 'Microsoft Corp.', 'Johnson & Johnson', 'JPMorgan Chase', 'Procter & Gamble']
model += lpSum(select[a] for a in stock_assets) <= 5, "Max_5_Stocks"

# Sector diversification: at least 3 distinct sectors
sector_map = {
    'Apple Inc.': 'Technology',
    'Microsoft Corp.': 'Technology',
    'Johnson & Johnson': 'Healthcare',
    'JPMorgan Chase': 'Financials',
    'Procter & Gamble': 'Consumer Goods',
    'Corporate Bond Fund': 'Fixed Income',
    'US Treasury 10-Year': 'Fixed Income'
}
sectors = set(sector_map.values())
sector_vars = {s: LpVariable(f"use_sector_{s.replace(' ', '_')}", cat=LpBinary) for s in sectors}
for a, s in sector_map.items():
    model += select[a] <= sector_vars[s], f"Sector_Link_{a}"
model += lpSum(sector_vars.values()) >= 3, "Min_3_Sectors"

# Bonds ≥ $250,000 (25%)
bond_assets = ['Corporate Bond Fund', 'US Treasury 10-Year']
model += lpSum(alloc[a] for a in bond_assets) >= 250_000, "Min_25_Bond_Allocation"

# Investment grade ≥ $600,000 (60%)
inv_grade_assets = [a for a in mean_returns if asset_char_df.loc[asset_char_df['Asset_Name'] == a, 'Is_Investment_Grade'].values[0]]
model += lpSum(alloc[a] for a in inv_grade_assets) >= 600_000, "Min_60_Investment_Grade"

# International assets ≥ $50,000 or $0 — and at least one must be selected
intl_assets = [a for a in mean_returns if asset_char_df.loc[asset_char_df['Asset_Name'] == a, 'Is_International'].values[0]]
for a in intl_assets:
    model += alloc[a] >= 50_000 * select[a], f"Intl_Min_50k_{a}"
model += lpSum(select[a] for a in intl_assets) >= 1, "AtLeast_One_International"

# Max 6 total assets
model += lpSum(select[a] for a in mean_returns) <= 6, "Max_6_Assets"

# Solve the model
model.solve()

# Output results
print("\n Optimal Portfolio Allocation:")
for a in alloc:
    if alloc[a].varValue > 1e-2:
        pct = 100 * alloc[a].varValue / TOTAL_INVESTMENT
        print(f"{a}: ${alloc[a].varValue:,.2f} ({pct:.2f}%)")

print(f"\n C1003 Total Expected Return: ${value(model.objective):,.2f} ({100 * value(model.objective) / TOTAL_INVESTMENT:.2f}%)")
```

#### 5.3.8 Multi-Objective Portfolio Optimization<a id="5.3.8-Multi-Objective-Portfolio-Optimization"></a>

- **Model Objective**:  
  This multi-objective portfolio optimization model is designed to balance two competing goals: maximizing expected return and minimizing portfolio variance. The model reflects investor preferences through a tunable parameter, `λ_risk`.

- **Key Configuration**:  
  - `λ_risk` ranges from 0 (pure return focus) to 1 (pure risk aversion)
    
  - The model uses a quadratic formulation to calculate total return and variance
    
  - The objective function is a convex combination:
     
    `Objective = λ * Variance - (1 - λ) * Return`
    
  - Asset weights are constrained to sum to 1, ensuring the portfolio is fully invested

- **Optimization Outcome (λ = 0.5)**:  
  - **Expected Return**: 3.52%
    
  - **Portfolio Variance**: 0.007754 (≈ 0.78%)
    
  - **Asset Allocation**:  
    - Apple Inc. (S001): 91.32%
       
    - Microsoft Corp. (S002): 8.68%

- **Strategic Insight**:  
  The λ = 0.5 configuration represents a balanced investor who values both growth and stability.  
  The optimizer concentrates most capital in Apple due to its high return-to-risk ratio, while still including Microsoft to reduce overall volatility.

```python
# Load asset return data
asset_returns_df = pd.read_csv("asset_returns.csv")
asset_returns_df['Date'] = pd.to_datetime(asset_returns_df[['Year', 'Month']].assign(DAY=1))

# Pivot the return data by Asset_ID
pivot_returns = asset_returns_df.pivot(index='Date', columns='Asset_ID', values='Return')
pivot_returns = pivot_returns.dropna(axis=1)  # Drop assets with missing values

# Calculate average return and covariance matrix
mean_returns = pivot_returns.mean()
cov_matrix = pivot_returns.cov()

# Asset list and count
assets = list(mean_returns.index)
n = len(assets)

# Trade-off parameter: 0 = focus on return, 1 = focus on variance
lambda_risk = 0.5

# Create Gurobi model
m = Model("Multiobjective_GoalProgramming")
m.setParam('OutputFlag', 0)

# Decision variables
w = {i: m.addVar(lb=0, ub=1, name=f"w_{i}") for i in assets}

# Objective: weighted return expression
return_expr = sum(w[i] * mean_returns[i] for i in assets)

# Objective: portfolio variance expression
variance_expr = QuadExpr()
for i in range(n):
    for j in range(n):
        variance_expr.add(w[assets[i]] * w[assets[j]] * cov_matrix.iloc[i, j])

# Objective function: maximize (1 - λ) * Return - λ * Variance
m.setObjective((1 - lambda_risk) * return_expr - lambda_risk * variance_expr, GRB.MAXIMIZE)

# Total weight constraint
m.addConstr(sum(w[i] for i in assets) == 1)

# Solve
m.optimize()

# Output
print(f"\nWeighted λ = {lambda_risk}")
print(f"Expected Return: {sum(w[i].X * mean_returns[i] for i in assets):.4f}")
print(f"Portfolio Variance: {sum(w[i].X * sum(w[j].X * cov_matrix.loc[i, j] for j in assets) for i in assets):.6f}")
print("Asset Allocation:")
for i in assets:
    if w[i].X > 0.001:
        print(f"{i}: {w[i].X:.4f}")
```

### EfficientFrontier

* This optimization model iterates over a range of λ (lambda) values from 0 to 1, where each λ represents a trade-off between maximizing expected return and minimizing portfolio variance. 
* For each λ, the code defines and solves a quadratic programming model that assigns weights to assets subject to full investment (weights sum to 1) and no short selling (weights between 0 and 1).
* The objective is a convex combination of return and risk, adjusted by λ.
* The Efficient Frontier's visualized line chart shows the relationship between portfolio risk (standard deviation) and expected return, which helps visualize the set of optimal portfolios under different risk preferences.


```python
from gurobipy import Model, GRB, QuadExpr
import numpy as np
import matplotlib.pyplot as plt


lambda_values = np.linspace(0, 1, 21)
returns = []
variances = []

for lam in lambda_values:
    m = Model("EfficientFrontier")
    m.setParam('OutputFlag', 0)

    w = {i: m.addVar(lb=0, ub=1, name=f"w_{i}") for i in assets}

    return_expr = sum(w[i] * mean_returns[i] for i in assets)

    variance_expr = QuadExpr()
    for i in range(n):
        for j in range(n):
            variance_expr.add(w[assets[i]] * w[assets[j]] * cov_matrix.iloc[i, j])

    m.setObjective((1 - lam) * return_expr - lam * variance_expr, GRB.MAXIMIZE)
    m.addConstr(sum(w[i] for i in assets) == 1)
    m.optimize()

    if m.Status == GRB.OPTIMAL:
        r = sum(w[i].X * mean_returns[i] for i in assets)
        v = sum(w[i].X * sum(w[j].X * cov_matrix.loc[i, j] for j in assets) for i in assets)
        returns.append(r)
        variances.append(v)

# Plot Efficient Frontier
plt.figure(figsize=(8, 5))
plt.plot(np.sqrt(variances), returns, marker='o', color='teal')
plt.title("Efficient Frontier (Expected Return vs Portfolio Risk)", fontsize=13)
plt.xlabel("Portfolio Risk (Std Dev)", fontsize=11)
plt.ylabel("Expected Return", fontsize=11)
plt.grid(True)
plt.tight_layout()
plt.show()
```
